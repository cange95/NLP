```python
def prepare_URL(ticker, N_items='50', type_='article', token_=secret, sources_list=",".join(sources_list), date = None):
    
    url = 'https://stocknewsapi.com/api/v1?tickers='
    if len(sources_list):

        url = str(url+ '&tickers='+ticker.upper() +'&items='+N_items + '&type='+type_ + '&token='+secret + '&sources='+sources_list)

    else:

        url = str(url+ '&tickers='+ticker.upper() +'&items='+N_items + '&type='+type_ + '&token='+secret )
    if date != None:
        url = str(url + '&date=' + date)

    print(url)
    return url
```

# NewAPI

```python
import pprint
import requests
```

```python
secret = '96d7019378434230a4cde7cb57aa7645'
```

```python
# Define the endpoint
url = 'https://newsapi.org/v2/top-headlines?'
#url = 'https://newsapi.org/v2/everything?'
```

```python
# Specify the query and number of returns
parameters = {
    'q': 'Apple', # query phrase
    #'sortBy':'publishedAt',
    'language':'en',
    'category':'business',
    #'sources':'reuters',
    'pageSize': 100,  # maximum is 100
    'apiKey': secret, # your own API key
    #'from':'2021-09-03',
    #'to':'2021-09-03'
}
```

```python
# Make the request
response = requests.get(url, params=parameters)

```

```python
# Convert the response to JSON format and pretty print it
response_json = response.json()
pprint.pprint(response_json)

```

```python
for num, article in enumerate(response_json['articles']):
    print(num, ')',' On ', article['publishedAt'], ' by ', article['source']['name'],'. Title: ', article['title'])
    print('\n')
```
